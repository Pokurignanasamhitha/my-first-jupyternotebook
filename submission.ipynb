```python
import pandas as pd 
df_meal=pd.read_csv("C:\\Users\\HP\\Desktop\\meal_info.csv")
df_meal.head()
```

```python

df_center=pd.read_csv("C:\\Users\\HP\\Desktop\\fulfilment_center_info.csv")
df_center.head()
```

```python
df_food=pd.read_csv("C:\\Users\\HP\\Desktop\\train.csv")
df_food.head()
```

```python
df3=pd.merge(df_food,df_center,on="center_id",how="outer")
df=pd.merge(df3,df_meal,how="outer",on="meal_id")
df
```

```python
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd
```

```python
table=pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)
table
plt.xlabel('Food Items')
plt.ylabel('Quantity Sold')
plt.title('Most Popular Food')
table.plot(kind="bar",width=0.8)
plt.xticks(rotation =70)
plt.show();
```

```python
df["revenue"]=df["checkout_price"]*df["num_orders"]
df['month']=(df['week']-1)//4
df
```

```python
a=pd.pivot_table(data=df,index='month',values='revenue',aggfunc=np.sum)
month=a.index.to_list()
month_order=a['revenue'].to_list()
b=pd.pivot_table(data=df,index='week',values='revenue',aggfunc=np.sum)
week=b.index.to_list()
week_order=b['revenue'].to_list()
fig, axes=plt.subplots(nrows=1,ncols=2,figsize=(20,5))
axes[0].plot(week,week_order)
axes[0].set_xlabel("Week")
axes[0].set_ylabel("Revenue")
axes[0].set_title("Weekly income")
axes[1].plot(month,month_order)
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Revenue")
axes[1].set_title("Monthly income")
plt.tight_layout(pad=3)
plt.show()
```
